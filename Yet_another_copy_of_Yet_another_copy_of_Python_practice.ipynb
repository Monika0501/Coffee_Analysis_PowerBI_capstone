```python
#python program to create a simple calculator

#3 steps to build calculator program
#1.functions for operations
#2.user input
#3.print result

#step 1 create functions
#function to add two numbers
def add(num1,num2):
  return num1+num2

#function to sub two numbers
def sub(num1,num2):
  return num1-num2

#function to multiply two numbers
def multiply(num1,num2):
  return num1*num2

#function to divide two numbers
def divide(num1,num2):
  return num1/num2

#function to avg two numbers
def avg(num1,num2):
  return (num1+num2)/2

#step-2: user input

print("please select a operation:\n" \
      "1.Addition\n"\
      "2.Subtraction\n" \
      "3.multiplication\n" \
      "4.Division\n" \
      "5.Average\n")
select = int(input("select a operation from 1,2,3,4,5:"))

number1=int(input("Enter first number:"))
number2=int(input("Enter second number:"))

#step-3: print result
if select ==1:
  print(number1,"+", number2,"=",\
        add(number1,number2))
elif select ==2:
  print(number1,"-", number2,"=",\
        sub(number1,number2))
elif select==3:
  print(number1,"*",number2,"=",\
        multiply(number1,number2))
elif select==4:
  print(number1,"/",number2,"=",\
        divide(number1,number2))
elif select==5:
  print("(", number1,"+",number2,")","/",2,"=",\
        avg(number1,number2))
else:
  print("invalid operation! please select again!")
```

##Tuples in Python

```python
#Create tuples
# 1.using parenthesis
my_tuple =(1,2,3)
print(my_tuple)
print(type(my_tuple))

my_tuple1=(1,"madhav",True,3.15)
print(my_tuple1)
print(type(my_tuple1))

#2.without Parenthesis
tuple1=1,2,3
print(tuple1)
print(type(tuple1))

#3.tuple constructor
tuple2=tuple((1,5,9))
print(type(tuple2))
print(tuple2)

list1=[1,2,3]
new_tuple=tuple(list1)
print(new_tuple)
print(type(new_tuple))

#4. create a single element

a=("a")
print(a)
print(type(a))

b=("b",)  #comma added
print(b)
print(type(b))

#access tuple - indexing
fruits=("apple","mango","cherry")
print(fruits[0])
print(fruits[-1])
```

```python
#tuple slicing
new_tuple=(10,20,30,40,50)
print(new_tuple[0:3:2])
```

```python
#tuple operations
#concatenate -join tuple
tuple1=(1,2,3)
tuple2=("a","b")
tuple3=tuple1+tuple2
print(tuple3)

#repetitive
tuple2 = ("apple","banana")*2
print(tuple2)

#checking in element in a tuple
tuple22=(1,2,3)
print(10 in tuple22)
print(1 in tuple22)

#tuple iteration
#for loop

fruits=("apple","mango","cherry")
for i in fruits:
  print(i)

  #while loop
i=0
while i<len(fruits):
    print(fruits[i])
    i+=1
```

```python
#tuple methods
color=("blue","green","orange","blue")

#count
print(color.count("green"))
print(color.count("blue"))

#index
print(color.index("green")) #it finds index number
print(color.index("blue"))

```

```python
#tuple functions
numbers=(1,2,3,4)

#len
print(len(numbers))

#sum
print(sum(numbers))

#min
print(min(numbers))

#max
print(max(numbers))

#sort
print(sorted(numbers)) #tuple is now list

a=sorted(numbers)
numbers_sorted=tuple(a)
print(numbers_sorted)
```

```python
#tuple pack and unpack

a="madhav"
b=21
c="Engineer"

print(a,b,c)

tuple_pack=a,b,c
print(tuple_pack)

name,age,profession=tuple_pack
print(name)
print(age)
print(profession)
```

```python
#tuple modification
tuple_numbers=(10,20,30)

#tuple_numbers[0]=100    #error

#how to mutate/modify tuple
list_numbers=list(tuple_numbers)
print(list_numbers)

list_numbers[0]=100
print(list_numbers)

tuple_numbers=tuple(list_numbers)
print(tuple_numbers)
```

```python
#set methods
#1.union-combine elements from 2 sets
set1={1,2,3}
set2={3,5,6}  #doesnt show duplicate in output(3 element is repeated)
union_set=set1.union(set2)
print(union_set)
```

```python
#union alternative
union_set2=set1|set2
print(union_set2)
```

```python
#2.Intersection - common elements
set1={1,2,3,4}
set2={3,4,5}
inter_set= set1.intersection(set2)
print(inter_set)

#intersection altetnative method
inter_set2=set1&set2
print(inter_set2)
```

```python
#3.Diference-element present in first set only but not in second set
set1={1,2,3,4}
set2={3,4,5}
diff_set=set1.difference(set2)
print(diff_set)

#diference alternative
diff_set2=set1-set2
print(diff_set2)
```

```python
#4.symmertic difference-elemnt in either set but not in bot
set1={1,2,3,4}
set2={3,4,5,6}
sdiff_set=set1.symmetric_difference(set2)
print(sdiff_set)

#symm diff alternative
sdiff_set2=set1 ^ set2
print(sdiff_set)
```

##Dictionary in Python

```python
#syntax
#my_dict={"key1":"value1","key2":"value2",....}

#method-1 create dictionary using curly braces
cohort={"course":"Python","Instructor":"Rishabh Mishra","Level":"Beginner"}
print(cohort)
print(type(cohort))
```

```python
#method-2 using dict()constructor
person=dict(name="madhav",age=20,grade="A")
print(person)
print(type(person))
```

```python
#method-3 using list of tuples
person2=dict([("name","madhav"),("age",20),("city","Hyd")])
print(person2)
print(type(person2))
```

```python
#Access dictionary values
student={
    1:"Class-x",
    "name":"madhav",
    "grade":"A",
    "city":"Hyd"
}
print(student)
print(type(student))
print(student["name"])
print(student["grade"])
print(student[1])
```

```python
#Dictionaru Methods
student={
    1:"Class-X",
    "name":"Madhav",
    "grade":"A",
    "city":"Hyd"
}

#keys
print(student.keys())

#Values
print(student.values())

#items
print(student.items())

#get
print(student["name"])
#print(student["email"])  #it will pop an error
print(student.get("name"))
print(student.get("email")) #it will not pop an error#
print(student.get("email","nahi hai")) #it will give the value
```

```python
#Add/modify items in dictionary
employee={
    "name":"monika",
    "age":25,
    "Profession":"AP specilist",
    "Salary":25000
}
print(employee)

#add item - used assing operator
employee["email"]='monika@example.com'

print(employee)

#modify/replace item
employee["Salary"]=30000
print(employee)

#remove items
#delete to remove item

del employee ["Profession"]
print(employee)

#pop method
var1=employee.pop("email")
print(var1)

print(employee)
```
